```python
x=iris_data.iloc[:,:4].values
x
```

```python
y=iris_data.iloc[:,-1].values
y
```

```python
sns.pairplot(data=iris_data,hue='target',kind='scatter',diag_kind='hist',palette=['red','green','blue'],markers=['o','d','s'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
train_score_list=[]
test_score_list=[]
k_range=range(1,16)
for k in k_range:
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train,y_train)
    y_pred_train=knn.predict(x_train)
    y_pred_test=knn.predict(x_test)
    accuracy1=accuracy_score(y_train,y_pred_train)
    accuracy2=accuracy_score(y_test,y_pred_test)
    train_score_list.append(accuracy1)
    test_score_list.append(accuracy2)
    print('k=%d accuraccy1=%.2f%%'%(k,accuracy1*100))
    print('k=%d accuraccy2=%.2f%%'%(k,accuracy2*100))
```

```python
plt.plot(k_range,train_score_list,color='red')
plt.plot(k_range,test_score_list,color='green')
plt.xlabel("k")
plt.ylabel("accuracy")
plt.title("k vs. accuracy")

plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
y_pred_train=knn.predict(x_train)
y_pred_test=knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train,y_pred_train))
print(accuracy_score(y_test,y_pred_test))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_train,y_pred_train)
print(cm)
cm1=confusion_matrix(y_test,y_pred_test)
print(cm1)
```

```python
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
sns.heatmap(cm1, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
