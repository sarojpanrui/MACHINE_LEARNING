```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([340,1080,640,880,990,510])
y = np.array([500,1700,1100,800,1400,500])

plt.scatter(x, y, color = "m",marker = "*",label='actual ouput')    

# number of observations/points
n = np.size(x)
 
# mean of x and y vector
m_x = np.mean(x)
m_y = np.mean(y)
 
# calculating cross-deviation and deviation about x
SS_xy = np.sum(y*x) - n*m_y*m_x
SS_xx = np.sum(x*x) - n*m_x*m_x
 
# calculating regression coefficients
b_1 = SS_xy / SS_xx
b_0 = m_y - b_1*m_x

print(b_1,b_0)

# predicted response vector
y_pred = b_0 + b_1*x
 
# plotting the regression line
plt.plot(x, y_pred, color = "g",label='regression line')

#prediction for unknown input
x=790
y_pred=b_0+b_1*x
print("the rent for the area 790 is %.2f"%y_pred)

plt.scatter(x, y_pred,color="red",marker="*",label='unknown input predicted output')
plt.text(x+20,y_pred+20,y_pred) 
# putting labels
plt.xlabel('area')
plt.ylabel('rent')
plt.title("house price")
plt.legend()
# function to show plot
plt.show()
```
