```python
import numpy as np
import matplotlib.pyplot as plt

X=np.array([340,1080,640,880,990,510])
Y=np.array([500,1700,1100,800,1400,500])

plt.scatter(X, Y, marker='*',color='b',label="actual output")



# Building the model
m = 0
c = 0

L = 0.000001  # The learning Rate
epochs = 10000  # The number of iterations to perform gradient descent

n = len(X) # Number of elements in X

# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = m*X + c  # The current predicted value of Y
    D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c
    
print (m, c)
#print("%.2f %.2f"%(m,c))
#print(f"the result is {m:.2f} and {c:.2f}")

# Making predictions
Y_pred = m*X + c


plt.plot(X,Y_pred, color='red',label="predicted output")  # regression line
X=790
Y_pred = m*X + c

plt.scatter(X,Y_pred, color='GREEN',marker="*")
plt.text(X+40,Y_pred+40,Y_pred)
plt.xlabel("area")
plt.ylabel("rent")
plt.title("house price")
plt.legend()
plt.show()
```
