```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1/(1+np.exp(-x))

def model(x,y,learning_rate,iterations):
    a1=0
    a0=0
    n=len(x)
    cost_list=[]
    for i in range(iterations):
        z=a1*x+a0
        a=sigmoid(z)
       
        #costfunction
        cost=-(1/n)*np.sum(y*np.log(a)+(1-y)*np.log(1-a))
        #gradient descent
        da1=(1/n)*np.sum((a-y)*x)
        da0=(1/n)*np.sum(a-y)
        a1=a1-learning_rate*da1
        a0=a0-learning_rate*da0
        #keeping track cost function value
        cost_list.append(cost)
        
    return a1,a0,cost_list
x= np.array([22, 25, 47, 52, 46, 56, 55,60,62,61 , 18,28,27,29,49 ])
y = np.array([0, 0, 1, 0, 1, 1, 0, 1, 1, 1, 0, 0,0,0,1])
n=len(x)
iterations=1000000
learning_rate=0.0001
a1,a0,cost_list=model(x,y,learning_rate,iterations)
print(a1,a0)
plt.scatter(x,y)
plt.xlabel("input")
plt.ylabel("output")
plt.title("logistic regression")
```

```python
plt.plot(np.arange(iterations),cost_list)
plt.xlabel('iteration')
plt.ylabel('cost')
plt.title('iteration vs cost')
plt.show()
```
